```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data-cleaning.csv')
```

```python
df.head(6)
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Salary'].iloc[0]
```

```python
df['Salary'] = df['Salary'].apply(lambda x:x.replace("$",''))
df['Salary'] = df['Salary'].apply(lambda x:x.replace("k",''))
```

```python
def average_of_this_two(args):
    try :
        return float(args)
    except:
        values = args.split('-')
        
        if len(values) == 2:
            
            return float(int(values[0]) + int(values[1]) ) /2
        

        
        
df['Salary'] = df["Salary"].apply(average_of_this_two)
```

```python
df['Rating'].isnull().sum()
```

```python
df['Rating'] = df["Rating"].replace(-1,np.nan)
```

```python
df['Rating'] = df["Rating"].fillna(df['Rating'].mean())
```

```python
df['Location'] = df["Location"].apply(lambda x:x.split(',')[0])
```

```python
df["Established"] = df["Established"].replace(-1,np.nan)
```

```python
df["Established"].dropna()
```

```python
df['Easy Apply']  =  df["Easy Apply"].replace("-1","FALSE")
```

```python
df.dropna()
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df['Age'] = df['Age'].round().astype(int)
df.head()
```

```python
df['Rating'] = df['Rating'].round()
df.head()
```

```python
df.to_csv('Cleaned-Data.csv', index=False)
```
